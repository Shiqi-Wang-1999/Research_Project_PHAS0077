```python
from selenium import webdriver
import time
```

```python

# Function of crawler, pass in the pages range to use it
def news_headlines_crawler(page_start, page_end):
    
    headlines = []
    dates = []
    
    browser = webdriver.Chrome(r"D:\Workplace\Jupyter\Research project\Oil-Price-Prediction\chromedriver.exe")
    
    for page in range(page_start, page_end):
        # there are many pages of the news
        url = f"https://www.investing.com/commodities/crude-oil-news/{page}"
        
        # Using chrome driver to access the url
        browser.get(url)

        # Finding news headlines and their corresponding dates
        elements= browser.find_elements_by_xpath("//*[@class='title' or @class='date']")
        elems = elements.copy()

        raw_data = []
        for post in elems:
            raw_data.append(post.text)

        # Delete the first three empty values
        del raw_data[:3]

        # Delete the last four useless values
        del raw_data[-4:]

        i=0
        for item in raw_data:
            if item == '':
                continue
            
            # Odd numbered items are news headlines text
            if i== 0 :
                headlines.append(item)
                i+= 1
            # Even numbered items are date text
            else :
                dates.append(item[3:])
                # One page includes 10 groups of news headlines and corresponding dates
                if len(dates) % 10 == 0:
                    break
                i-= 1
        
    return (headlines, dates)
```

```python
import pandas as pd

# call the function 'news_headlines_crawler'
headlines_dates_tuple = news_headlines_crawler(3, 4)

# construct the dictionary
dic = {'headlines' : headlines_dates_tuple[0], 'dates': headlines_dates_tuple[1]}

# construct the dataframe
df_news = pd.DataFrame(data=dic)
df_news
```

```python
# Store the data
df_news.to_excel('Oil_News_Title_Date.xlsx', index = False)
```
